```python
f = open("words.txt", "r", encoding="utf-8")
words = f.read().split("\\\\n")
words
```

```python
word_list = dict()

for i in range(len(words)):
    if(":" in words[i]):
        word_list[words[i].split(":")[0]] = dict()
        word_list[words[i].split(":")[0]]["meaning"] = words[i].split(":")[1]
        j = i+1
        sentence = ""
        while(words[j][-1]!= "."):
            sentence += words[j]
            j += 1
            sentence += " "
        sentence += words[j]
        word_list[words[i].split(":")[0]]["example"] = sentence
```

```python
with open("vocab.json", "w", encoding="utf-8") as f:
    f.write(str(word_list).replace("\\\\'", "\\\\""))
```

```python
# scrape pdf to get the words

import PyPDF2

pdf = open("graduateshotline_word_list.pdf", "rb")

pdf_reader = PyPDF2.PdfReader(pdf)
```

```python
# extract text from pdf

text = ""

for i in range(len(pdf_reader.pages)):
    text += pdf_reader.pages[i].extract_text()

text = text.split("\\\\n")
text
```

```python
words = dict()
for line in text:
    if("  " in line):
        word_meaning = line.split("  ")
        if("www." in word_meaning[0]):
            continue
        if(len(word_meaning)>1):
            if(word_meaning[1]):
                words[word_meaning[0].strip()] = dict()
                words[word_meaning[0].strip()]["meaning"] = word_meaning[1].strip()
```

```python
from bs4 import BeautifulSoup

# parse test.html file and extract the words

with open("test.html", "r", encoding="utf-8") as f:
    soup = BeautifulSoup(f, 'html.parser')

# extract words from "wordlist words notesView" class

words = soup.find_all("a", class_="word")
definitions = soup.find_all("div", class_="definition")
```

```python
# map ith word to ith definition

word_list = dict()

for i in range(len(words)):
    curr_word = words[i].text.strip()
    word_list[curr_word] = dict()
    word_list[curr_word]["meaning"] = definitions[i].text

word_list
```

```python
df = pd.read_csv("vocab_combined.csv")

# find all examples that has ___ as substring then replace it with the word

import re

df.columns = ["word", "meaning", "example"]
for i in range(len(df)):
    example = df.iloc[i].example
    word = df.iloc[i].word
    if("________" in example):
        example = re.sub("________", word, example)
        df.iloc[i].example = example
    if('“' in example):
        example = re.sub('“', '"', example)
        example = re.sub('”', '"', example)
        df.iloc[i].example = example

# find all examples that has ___ as substring

df[df.example.str.contains("“")].shape[0]

```
